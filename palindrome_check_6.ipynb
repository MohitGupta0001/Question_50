```python
def check_palindrome(string:str)->str:
    """
    Return palindrome string.
    Args:Take only string.
    Return:Return palindrome string.
    """
    if string.strip() == "":
        raise TypeError ("String can not be take space and empty string.")
    if not isinstance(string,str) :
        raise TypeError("Take only string not any thing else.")
    rev=""
    for i in string:
        rev=i+rev
    if rev==string:
        print(f"Its palindrome string {rev}")
    else:
        print(f"Its  not palindrome string {rev}")
if __name__ == "__main__":
    try:
        string = input("Enter a string:")
        check_palindrome(string)
    except (TypeError,ValueError) as e:
        print("Error:",e)
    except Exception as e:
        print(e)
    finally:
        print("Program run successfully.")
```
