```python
def reverse_string(string: str)-> str:
    """
    Args:
        Take only string input.
    Returns:
        return given input string reverse.
    """
    
    reverse=""
    if not isinstance(string,str):
        raise TypeError ("Only take input string not any thing it will be taken.")
    if string.isdigit():
        raise ValueError("Digits is not allow in input.")
    else:
        for i in string:
            reverse=i+reverse
        print(f"Reverse string:{reverse}")
if __name__ == "__main__":
    try:
        user_input=input("Enter a string:")
        reverse_string(user_input)
    except (TypeError,ValueError) as e:
        print("Error:",e)
    finally:
        print("program run successfully.")
        
    
```
