```python
def is_prime_number(number):
    """
    Purpose: Check whether a given number is prime or not
    Argument:
        number (int) → Input number
    Return Value:
        True  → Number is prime
        False → Number is not prime
    """
    if number <= 1:
        return False
    for i in range(2, number):
        if number % i == 0:
            return False
    return True
try:
    num = int(input("Enter a number: "))
    result = is_prime_number(num)
    if result:
        print("Prime number")
    else:
        print("Not a prime number")

except ValueError as e:
    print("Error: Please enter a valid integer number only.")
except Exception as e:
    print("Something went wrong:", e)
finally:
    print("Your program run successfully.")
```
