```python
def calculate_factorial(num:int)->int:
    """
    Return the calculate the number of factorial.
    Args:Take number only.
    Return:return number factorial calculate number.
    """
    if not isinstance(num,int):
        raise TypeError("Take only numbers,please enter number.")
    if num==0:
        raise ValueError("Number must greater than 0.")
    if num.strip() == "" or not num.isdigit():
        raise ValueError("Enter a valid number (not empty or space).")
    fact=1
    for i in range(1,num+1):
        fact=fact*i
    return fact
if __name__ == "__main__":
    try:
        num=input("Enter a number:")
        calculate_factorial(num)
    except TypeError as e:
        print("Error:",e)
    except ValueError as e:
        print("Error:",e)
    except Exception as e:
        print(e)
    finally:
        print("Program run succesfully.")
```
