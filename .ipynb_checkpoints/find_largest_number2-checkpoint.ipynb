```python
def find_largest_number(nums:str):
    """
    Args:
        nums (str):take number with space separated.
    Returns:
        return largest number.
    """
    if not nums.strip():
        raise ValueError("Input cannot be empty")
    list1 = nums.split()
    for i in list1:
        if not i.isdigit():
            raise TypeError("Only enter numeric input.")
    largest_number=int(list1[0])
    for i in list1:
        if int(i)>largest_number:
            largest_number=int(i)
    print("Largest number is:",largest_number)
if __name__ == "__main__":
    try:
        nums=input("Enters numbers:")
        find_largest_number(nums)
    except (TypeError,ValueError) as e:
        print("Error:",e)
    finally:
        print("program run successfully.")
    
```

```python
def find_largest_number(list1):
    """
    Args:
        Take only numbers.
    Returns:
        return largest number.
    """
    largest_number = list1[0]
    
    for i in list1:
        if i > largest_number:
            largest_number = i
            
    return largest_number


if __name__ == "__main__":
    try:
        numbers = input("Enter numbers separated by space: ")
        list1 = [int(x) for x in numbers.split()]
        
        result = find_largest_number(list1)
        print("Largest number is:", result)
    finally:
        print("program run successfully.")
```

```python
def fln():
    """
    Take only number.
    Args:Take input only number.
    Return:return laresgt number.
    """
    n=int(input("Enter numbers:"))
    if n<=0:
        raise ValueError("Number must be greater than 0.")
    list1=[]
    for i in range(n):
        value=input(f"Enter number:{i+1}")
        if value.strip() == "":
            raise ValueError("gave a number not take empty.")
        if not value.lstrip().isdigit():
            raise ValueError("Only nubmeric value allowed.")
        num=int(value)
        list1.append(num)
        largest_number=list1[0]
    for num in list1:
        if num>largest_number:
            largest_number=num
    print("Largest number:",largest_number)
if __name__ == "__main__":
    try:
        fln()
    except ValueError as e:
        print("Error:",e)
    finally:
        print("program run sucessfully.")
        
```
