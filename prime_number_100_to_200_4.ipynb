```python
def prime_number(start, end):
    """
    Return the prime number start to end.
    Args=Take number to start to end:
    Return=retrun the all prime which are under the number of the range.
    """
    for num in range(start, end + 1):
        if num < 2:
            continue
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num)
if __name__ == "__main__":
    try:
        prime_number(100, 200)
    except Exception as e:
        print(e)
    finally:
        print("Program run successfully.")

```
